# Matplotlib -də legend sinifi 

<p style="font-family:Times New Roman; font-size:166em;">&emsp;&emsp;Legend qrafik göstəriciləri ifadə edən açıqlamadır. <br>O, matplotlib.legend modulunun alt sinifi olan Legend -i çağırar.</p><br>İlk öncə rəsmi dokumentasiyada sizə lazım ola biləcək 4 anlayışı izah edək:

><i><b>Legend entry:</b></i><br>

&emsp;&emsp;Bir <b>legend</b> (açıqlama) bir və ya bir neçə giriş -entry dən ibarət ola bilər. <br>Bir entry yalnız və yalnız bir <i>legend key</i> -indən və bir <i>legend label</i> -indən <br>ibarət ola bilər.

><i><b>Legend key:</b></i><br>

&emsp;&emsp;Qrafikin stilini fiqurla açıqlamanın solunda göstərir. Məsələn, xətti <br>qrafikdə uluzlu,  nöqtəli xətt, qırıq-qırıq xətt və s.

><i><b>Legend label:</b></i><br>

&emsp;&emsp;Legend key -i ilə daxil edilmiş Legend handle (işarə (bir növ marker)) <br>-sinin qrafikdə nəyi ifadə etdiyini açıqlayar.
><i><b>Legend handle:</b></i><br>

&emsp;&emsp;Legend- də müvafiq giriş qrafiki üçün istifadə edilən orijinal işarə - <br>simvol (və ya marker) i ifadə edər.
