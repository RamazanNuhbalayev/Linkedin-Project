<i><b>format strings:</b></i> bunlar string formasında yazdığımız formatlama qısaltmalarıdır.

Daxil edilmədikdə, cycle stilindən olan dəyər istifadə olunur. İstisna: Əgər line -nin formatı verilirsə, lakin marker -in dəyəri daxil edilməzsə, data markersiz xətt ilə ifadə olunacaq. [rəng][marker][xətt] kimi digər kombinasiyalar da daxil edilə bilər, lakin verilən string nəticəsində ekrana gələn qrafik vizualın nəticəsində hansı string -in nəyi təyin etdiyini anlamaqda çətinlik çəkə və hətta anlamaya bilərsiniz.Bu linkdə daha müfəssəl açıqlamaları oxuya bilərsiniz: <br><a href = 'https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html'>https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html</a>

<i><b>plt.plot()</b></i>: 
> Qrafiki yaradan əsas əmrdir. Bu əmr ilk əvvəl  x oxu üzrə , ardınca da y oxu üzrə parametrləri alar. Şəxsi tövsiyyəm odur ki, bu parametrləri daxil edərkən, <i>'x='</i> , <i>'y='</i> və <i>'data='</i>  formadasında mənimsədəsiniz. format stringi daxil edməklə sadə vizual dəyişiklikləri (qrafikin rəngi, tipi(qırıq-qırıq,nöqtəli, müsbət işarəli və s.), işarələyici simvol) yerinə yetirə bilərsiniz.<br> Lakin daxilindəki 3 parametrdən istifadə etməklə də bunları təyin edə bilərsiniz: 
1. <i>color</i> - rəngi
2. <i>linestyle</i> - xətt stilini
3. <i>marker</i> - işarələyiciləri <br>
bildirər.

<i><b>figsize</b></i>:
> Grafikin çərçivə ölçüsünü - nə qədər yer tutacağını təyin edərik. Ölçülər <i>'inç'</i> uzunluq ölçü vahidi ilə ifadə edilər. Defolt olaraq təyin edilmiş ölçülər [6.4, 4.8] [en, hündürlük] -dir. .İstifadə zamanı adətən  plt.figure(figsize=(8,6)) təyin edilər.<br>

<i><b>plt.title()</b></i>: 
> Ümumi qrafikə başlıq - label təyin edər.<br>Daxilindəki <i>loc=</i> parametri başlığın sol, orta və ya sağ tərəfdən (küncdən) yerləşəcəyini müəyyənləşdirər - təyin edər.

<i><b>plt.xlabel()</b></i>: 
> x oxuna başlıq - label təyin edər.<br>Daxilindəki <i>loc=</i> parametri başlığın sol, orta və ya sağ tərəfdən (küncdən) yerləşəcəyini müəyyənləşdirər - təyin edər.

<i><b>plt.ylabel()</b></i>: 
> y oxuna başlıq - label təyin edər.<br>Daxilindəki <i>loc=</i> parametri başlığın sol, orta və ya sağ tərəfdən (küncdən) yerləşəcəyini müəyyənləşdirər - təyin edər.

<i><b>plt.show()</b></i>: 
> Qrafiki vizuallaşdırmağımıza imkan tanıyar. Belə ki, kod blokda verdiyimiz müxtəlif açıq qrafiki fiqurları bizə geri döndərər. Grafikləri ekrana vermək üçün bəzən onun funksiyası ilə çağırmaq yetərli olsa da bu müəyyən hallarda mümkün olmur. Ona görə də plt.show() istifadəsi daha məqsədə uyğundur.<br> 

&emsp;&emsp;İndi gəlin, bu kodlara nümunə üzərindən baxaq:

&emsp;&emsp;ilk öncə sadə bir hesabdan bəhs edim: 1 pixel = 0.010417 inç .  İnç vahidi ilə işləmək narahatlıq verir və biz ekran ölçülərini piksellə bilirk ona görə bu problemi kökündən həll edək və funksiya yaradaq:


```python
def pti (pti_X, pti_Y):
   '''
   İki ədəd pixel dəyər alar və onları 
   inç uzunluq vahidinə çevirib geri qaytarar
   '''  
   x_Res = pti_X*0.010417
   y_Res = pti_Y*0.010417
   return x_Res, y_Res
```

&emsp;&emsp;Docstringdə dediyim kimi iki dəyişən alar və onları tuple formasında geri qaytarar. Biz də bunu asanlıqla figsize parametrinə <br>
> figsize = pti (En_piksel_İlə, Hündürlük_piksel_İlə) 

bu formada tətbiq edərik.

```python
from pandas.core.frame import fmt
plt.figure(figsize=pti(1800,840))

plt.subplot(2,1,1)
plt.title(label='Meşə ehtiyatlarının bərpası ilə bağlı və yanğınların\
 əhatə etdiyi ərazi göstəricilərinin qarşılıqlı müqayisəsi', loc='center')
plt.plot(df.index, 'Meşə əkini və səpini  (min hektar)', 
         color='grey',linestyle='--', marker='o',data=df)
plt.plot(df.index, 'Meşənin təbii bərpasına yardım  (min hektar)',
         '-.b',data=df)
plt.plot(df.index, 'Meşə yanğınında meşə ilə örtülü sahə  (min hektar)', 
         data=df, color = 'orange')
plt.xlabel(xlabel='Illər', loc='center')
plt.ylabel(ylabel='Ərazi ölçüsü (min hektar)', loc='center')
plt.xticks(range(2000,2022))
plt.yticks(range(14))
plt.legend()
plt.subplots_adjust(bottom=-0.0005)

plt.subplot(2,1,2)
plt.title(label='Meşə yanğınlarında sağlam və ya dik\
 vəziyyətdə kökü üstə yanmış ağacların həcm göstəriciləri ', loc='center')
plt.plot(df.index, 'Meşə yanğınında kökü üstündə\
 yanmış ağacların həcmi  (kub metr)',data=df)
plt.xlabel(xlabel='Illər', loc='center')
plt.ylabel(ylabel='Ağacların həcmi  (kub metr)', loc='center')
plt.xticks(range(2000,2022))
plt.yticks(range(0,480,30))

plt.show()
```
