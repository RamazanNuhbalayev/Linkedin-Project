&emsp;&emsp;İlk öncə məqsədimizi qeyd edək ki, hərşey anlaşılan olsun. 3 sadə amma faydalılıq baxımından çox yüksək keyfiyyətli funksiya vardır: <br> 
1.   head ( ) - bu funksiya bizə data haqqında anlayış verməsi baxımından defolt olaraq onun ilk  5 sətirini , bütun sütunlarla qarşımıza çıxardır və data haqqında ilkin biliklərə malik oluruq. Bu funksiyadan istifadədə yaranan çətin problemlərdən biri də SQL və digər DMS (Database management systems) -lərdə heç bir dataya müdaxiləyə izn olmamasından yaranır. Belə ki, test mühitində işləyən verilənlər bazaları ilkin məlumatları sınaq üçün yazılmış və birçox hissəsi düzgün göstəricilərlə əhatə olunmamış ola bilir belə ki, bir çox hallarda verilənlər bazasınıın test mühitinin belə toxunulmazlığı qaydaları (işləri idarə edən şirkət, qurum və mövcud ölkənin normativ hüquqi aktlarından irəli gələn tələblər) bu data -ları silməyə izn vermir. &emsp;&emsp;<i><b>index</b></i> bizə defolt olaraq əgər indeks təyin edilməzsə  index aralıqlarını bizə göstərər (başlanğıc, bitiş, addım) - formasında<br>&emsp;&emsp;Yox əgər index dataframe yaradılarkən təyin edilmişdirsə , təyin edilmiş array (sütun) -u bizə geri döndürər.<br>&emsp;&emsp;<i><b>index.names</b></i> index -ə təyin etdiyimiz sütun adını qaytarar.

Parametrlər<br><br>
by : str və ya str -lardan ibarət list

>Sıralamaq üçün ad və ya adlar siyahısı qəbul edər.<br>
- if `axis`  0  və ya `'index'` -dirsa o zaman `by` index level -ləri və ya sütun başlıqlarını alar column labels.<br>
- if `axis` is 1  və ya `'columns'` -dirsa o zaman `by` column level -ləri və ya index başlıqları alar.<br>

axis : {0 ovə ya 'index', 1 və ya 'columns'}, default 0<br>
>Sıralamaq üçün axe və ya axislər verilər.<br>

ascending : bool və ya bool -lardan ibarət list, defolt True<br>
>Ascending və ya descending üzrə sıralaya bilərsiz. Çoxölçülü sıralama üçün bool -lardan ibarət list verilər. Amma, bu list by ilə göndərilən siyahı ilə eyni sayda element saxlamalıdır.<br>

inplace : bool, defolt False<br>
>Əgər True təyin edilərsə, əməliyyat in-place, yəni yerində - anındaca həyata heçirilir.<br>

kind : {'quicksort', 'mergesort', 'heapsort', 'stable'}, defolt 'quicksort'<br>
>Sıralama alqoritmi seçilir. Daha çox informasiya üçün həmçinin bax :funksiya:`numpy.sort`.<br> &emsp;&emsp;`mergesort` və `stable` yeganə stabil alqoritmlərdir. Dataframe -lərdə bu seçim sıralama bir sütun və ya  label üzərində tətbiq edilərkən istifadə edilir.<br>

na_position : {'first', 'last'}, default 'last'<br>
>Əgər `first`olarsa NaN dəyərləri başlanğıcda ; `last` olarsa NaNs dəyərləri cədvəl sonunda geri döndürər.<br>

ignore_index : bool, default False<br>
>Əgər True olarasa, nəticədə index `0, 1, …, n - 1` olaraq başlıqla sətirlərdə görünəcəkdir. `n` - burada datadakı sətirlərin sayıdır.<br>

Qaytarır:
>Sıralanmış obyekt : DataFrame və ya None. Sıralanmış dəyərlərdən ibarət DataFrame və ya None əgər ``inplace=True`` dəyəri olarsa.

## Pivot Table (Yekunlar Cədvəli) <br>
&emsp;&emsp;Əgər Exceli bilirsinizsə burda çəkinməyə heç bir əsas yoxdur. Çünki, buradakı anlayışlar tamamilə eynidir. Əgər bilmirsinizsə çox da narahat olmayın. Çünki istifadəsi çox sadədir.

```python
df_2 = pd.DataFrame({'Aylar': ['January','February','March','January','February',
                             'March','January','February','March'], 
                     'Ştat':['New York','New York','New York','Texas','Texas',
                            'Texas','Washington','Washington','Washington'], 
                     'Nəmlik': [20,25,65,34,56,85,21,56,79]})
df_2
```

&emsp;&emsp;Başlamazdan öncə corr() metodu haqqında ilkin fikir formalaşdıraq və aşağıdakı kodda onu çalışdıraq, bu metod bizə korrelasiya əlaqəsi haqqında fikir verir:

```python
df_2.corr()
```

&emsp;&emsp;İndi isə, 3 ədəd faiz nisbəti ilə ifadə edilmiş 3 ay üzrə 3 müxtəlif ştatdakı Nəmlik miqdarını Yekunlar cədvəlində ifadə edək.

```python
df_2.pivot_table(index = 'Aylar', columns = 'Ştat', values = 'Nəmlik')
```

```python
df_2.pivot_table(index = 'Ştat', columns = 'Aylar', values = 'Nəmlik')
```
